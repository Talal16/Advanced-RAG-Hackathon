```python
import os

documents = SimpleWebPageReader(html_to_text=True).load_data(
    [ "https://drive.google.com/file/d/1tTi9LOXHughlAIvSM7DRDCIJqGFrWD9B/view?usp=drive_link",
  "https://drive.google.com/file/d/1atiBEqVbydNitbufR6cmhD85GFzAerR-/view?usp=drive_link",
  "https://drive.google.com/file/d/1m4NNX05Q63GTjw-cZEc0V8HbB7j9Sks7/view?usp=drive_link"]
)
print(f"documents loaded into {len(documents)} document objects")
print(f"Document ID of first doc is {documents[0].doc_id}")
```

```python
index = VectaraIndex.from_documents(documents)
```

```python
from llama_index.core import PromptTemplate
system_prompt = system_prompt = "You are a lawyer assistant . Your goal is to answer questions as accurately as possible and explain answer to lawyers based on your training data and the instructions."
# This will wrap the default prompts that are internal to llama-index
query_wrapper_prompt = PromptTemplate("<|USER|>{query_str}<|ASSISTANT|>")
```

```python
llm = HuggingFaceLLM(
    context_window=2048,
    max_new_tokens=256,
    generate_kwargs={"temperature": 0.1, "do_sample": False},
    system_prompt=system_prompt,
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    model_name="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    device_map="cuda",
    # uncomment this if using CUDA to reduce memory usage
    model_kwargs={"torch_dtype": torch.bfloat16}
)
Settings.chunk_size = 512
Settings.llm = llm
```

```python
query = "What are some of the key features of the British legal system?"
```

```python
query_engine = index.as_query_engine(
    similarity_top_k=5,
    summary_enabled=False,
    llm=llm
)
response = query_engine.query(query)
print(response)
```

```python
def predict(input, history):
  response = query_engine.query(input)
  return str(response)
```

```python
import gradio as gr

gr.ChatInterface(predict).launch(share=True)
```
